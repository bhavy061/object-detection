```python
import cv2
import numpy as np
import tkinter as tk
import speech_recognition as sr
import pyttsx3
from cvzone.HandTrackingModule import HandDetector
from cvzone.FaceDetectionModule import FaceDetector

class StyledButton(tk.Button):
    def __init__(self, *args, **kwargs):
        tk.Button.__init__(self, *args, **kwargs)
        self.config(
            relief=tk.FLAT,
            bg="#3498db",
            fg="white",
            activebackground="#2980b9",
            activeforeground="white",
            padx=10,
            pady=5,
            font=("Helvetica", 12),
        )

def face_detection():
    sap = cv2.VideoCapture(0)
    detector = FaceDetector()
    while True:
        success, img = sap.read()
        img,bboxs = detector.findFaces(img)
        cv2.imshow("Image", img)
        if cv2.waitKey(1) == 13:
            break
    cv2.destroyAllWindows()

def fingerpoint_detection():
    ca = cv2.VideoCapture(0)
    model = HandDetector()
    while True:
        st, ph = ca.read()
        hand = model.findHands(ph)
        cv2.imshow("hii", ph)
        if cv2.waitKey(1) == 13:
            break
    cv2.destroyAllWindows()

def vehicle_detection():
    cap = cv2.VideoCapture("videop.mp4")
    min_width_rect = 80
    min_height_rect = 80
    algo = cv2.bgsegm.createBackgroundSubtractorMOG()
    while True:
        ret, frame1 = cap.read()
        grey = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)
        blur = cv2.GaussianBlur(grey, (3, 3), 5)
        img_sub = algo.apply(blur)
        dilat = cv2.dilate(img_sub, np.ones((5, 5)))
        kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
        dilatada = cv2.morphologyEx(dilat, cv2.MORPH_CLOSE, kernel)
        dilatada = cv2.morphologyEx(dilatada, cv2.MORPH_CLOSE, kernel)
        countershape, h = cv2.findContours(dilatada, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
        for (i, c) in enumerate(countershape):
            (x, y, w, h) = cv2.boundingRect(c)
            validate_counter = (w >= min_width_rect) and (h >= min_height_rect)
            if not validate_counter:
                continue
            cv2.rectangle(frame1, (x, y), (x+w, y+h), (0, 0, 255), 2)
        cv2.imshow("detector", frame1)
        if cv2.waitKey(1) == 13:
            break

    cv2.destroyAllWindows()
    cap.release()
def assistance():
    r = sr.Recognizer()
    flag=1
    while(flag):
        
    # Exception handling to handle
    
        try:

            with sr.Microphone() as source2:

            
                r.adjust_for_ambient_noise(source2, duration=0.2)
                print(" ðŸ¤– Hii i am your robo //")
                print("What can i do for you ")

                audio2 = r.listen(source2)


                MyText = r.recognize_google(audio2)
                MyText = MyText.lower()
                print()
                print("Did you say ",MyText)
                ch=MyText
                if "face detector" in ch:
                    face_detection()
                    print("Sure..")
                elif "hand detector" in ch:
                    fingerpoint_detection()
                    print("Sure..")
                elif "vehicle detection" in ch:
                    vehicle_detection()
                    print("Sure..")
                else:
                    print("i did'nt get that**//")
        except sr.RequestError as e:
            print("Could not request results; {0}".format(e))

        except sr.UnknownValueError:
            print("unknown error occurred")
        flag=flag-1

def on_button_click(choice):
    if choice == "1":
        face_detection()
    elif choice == "2":
        fingerpoint_detection()
    elif choice == "3":
         vehicle_detection()
    elif choice == "4":
        assistance()    
    else:
        print("invalid choice")
    
root = tk.Tk()
root.title("OBJECT DETECTOR")
root.config(bg="#2c3e50") 
# buttons with the StyledButton class
button_labels = ["Open FACE DETECTOR", "Open HAND DETECTOR", "Open VEHICLE DETECTOR","VOICE ASSISTANCE"]
buttons = []
for i, label in enumerate(button_labels):
    button = StyledButton(root, text=label, command=lambda idx=i: on_button_click(str(idx + 1)))
    buttons.append(button)
intro_label = tk.Label(root, text="Welcome!", fg="#2c3e50", font=("Helvetica", 14, "bold"))
intro_label.pack(pady=10)


# Pack buttons
for button in buttons:
    button.pack(fill=tk.BOTH, padx=10, pady=5)

frame = tk.Frame(root)
frame.pack(padx=20, pady=20)    
choice_var = tk.IntVar()
# the main loop
root.mainloop()
```
